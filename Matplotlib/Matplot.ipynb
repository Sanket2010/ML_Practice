# Bar Plot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x =["python","c","c++","java"]
y = [85,70,60,82]
plt.xlabel("Languages",fontsize = 15)
plt.ylabel("Poupularity",fontsize = 15)
plt.title("Programming")
c = ["y","b","m","g"]
plt.bar(x,y,width = 0.4,color=c,align = "center",edgecolor = "r", linewidth = 2,linestyle = ":",alpha = 0.4)
plt.show()
```

```python
x =["python","c","c++","java"]
y = [85,70,60,82]
plt.xlabel("Languages",fontsize = 15)
plt.ylabel("Poupularity",fontsize = 15)
plt.title("Programming")
c = ["y","b","m","g"]
plt.bar(x,y,width = 0.4,color=c,align = "center",edgecolor = "r", label = "Programming")
plt.legend()
plt.show()
```

```python
x =["python","c","c++","java"]
y = [85,70,60,82]
z = [20,30,40,50]

width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("Languages",fontsize = 15)
plt.ylabel("Poupularity",fontsize = 15)
plt.title("Programming")

plt.bar(p,y,width,align = "center", label = "Popularity")
plt.bar(p1,z,width,color="y",align = "center", label = "Popularity1")

plt.xticks(p+width/2,x,rotation = 10)
plt.legend()
plt.show()
```

```python
x =["python","c","c++","java"]
y = [85,70,60,82]
z = [20,30,40,50]

width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.ylabel("Languages",fontsize = 15)
plt.xlabel("Poupularity",fontsize = 15)
plt.title("Programming")

plt.barh(p,y,width,align = "center", label = "Popularity")
plt.barh(p1,z,width,color="y",align = "center", label = "Popularity1")

plt.xticks(p+width/2,x,rotation = 10)
plt.legend()
plt.show()
```

# Scatter Plot


```python
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,7,9]

colors = ["r","y","b","m","g","r","b"]
plt.title("Scatter",fontsize= 15)
plt.xlabel("Day",fontsize =15)
plt.ylabel("No.",fontsize = 15)
plt.scatter(day,no, c=colors)
plt.show()
```

```python
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,7,9]

colors = ["r","y","b","m","g","r","b"]
sizes = [400,200,300,100,500,600,100]

plt.scatter(day,no, c=colors,s = sizes, alpha = 0.4,marker="*", edgecolor ="g", linewidth=4)
plt.title("Scatter",fontsize= 15)
plt.xlabel("Day",fontsize =15)
plt.ylabel("No.",fontsize = 15)

plt.show()
```

```python
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,7,9]

colors = ["r","y","b","m","g","r","b"]
sizes = [400,200,300,100,500,600,100]

plt.scatter(day,no, c=colors,s=sizes)
plt.title("Scatter",fontsize= 15)
plt.xlabel("Day",fontsize =15)
plt.ylabel("No.",fontsize = 15)

plt.show()
```

```python
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,7,9]
no2 = [3,5,7,1,9,3,2]

colors = [10,40,30,49,69,29,59]
sizes = [400,200,300,100,500,600,100]

plt.scatter(day,no, c=colors,s=sizes, cmap = "viridis",alpha = 0.4)
plt.scatter(day,no2, c=colors,s=sizes, cmap = "viridis",alpha = 0.4)
t = plt.colorbar()
t.set_label("Color Bar",fontsize= 15)
plt.title("Scatter",fontsize= 15)
plt.xlabel("Day",fontsize =15)
plt.ylabel("No.",fontsize = 15)

plt.show()
```

# Pie Plot

```python
x = [10,20,30,40]
y = ["c","c++","Java","Python"]
ex = [0.4,0.0,0.0,0.0]
c = ["r","y","g","b"]

plt.pie(x,labels=y,explode = ex, colors = c, autopct="%0.1f%%" )




plt.show()
```
